```python
# there is one exta special character so we will remove those things by using function and regex
import re

def clean_name(names):
    if re.findall(r'[a-zA-Z0-9 ]',names):
        n=re.findall(r'[a-zA-Z0-9 ]',names)
        n1=''.join(n)
        return n1
    else:
        return names
```

```python
# changing data type 
df['Name']=df['Name'].astype(str)
```

```python
# applying the func clean_name 
df['Name']=df['Name'].apply(clean_name)
```

#### iv) Rating

```python
# checking in ratings column

df['Rating'].unique()
```

```python
# remving all NEW,nan,- values from rating column

def clean_rates(rate):
    if (rate == 'NEW' or rate == '-'):
        return np.nan
    else:
        rates=float(str(rate).split('/')[0])
        return rates
    
```

```python
# Applying the above function in Rating column
df['Rating']=df['Rating'].apply(clean_rates)
```

## 1.  List of Most expensive resturants 

```python
top_rated_res=df.groupby('Name')[['Cost for 2 People(approx)']].mean().sort_values(by='Cost for 2 People(approx)',ascending=False).head(10).reset_index()
top_rated_res
```

## 14. which cuisines charges more for two people ?

```python
cu_co = df.groupby('Available Cuisines')[['Cost for 2 People(approx)']].mean().sort_values(by='Cost for 2 People(approx)',ascending=False).reset_index().head(20)

plt.figure(figsize=(14,10))
c=sb.barplot(data=cu_co,x="Available Cuisines",y="Cost for 2 People(approx)",palette="flag_r",edgecolor="black",n_boot=1,alpha=0.7)
plt.xlabel("Available Cuisines Name",fontname="Times New Roman",fontsize=14,weight="bold")
plt.ylabel("Cost for 2 People(approx)",fontname="Times New Roman",fontsize=14,weight="bold")
plt.xticks(weight="bold",rotation=80)
plt.yticks(fontsize=18,weight="bold")
for i in c.containers:
    c.bar_label(i)
plt.show()
```

## 15. What is the Minimum Range for cuisines which charge less for two people ?

```python
cu_co_low = df.groupby('Available Cuisines')[['Cost for 2 People(approx)']].min().sort_values(by='Cost for 2 People(approx)',ascending=False).reset_index().tail(10)
cu=sb.barplot(data=cu_co_low,x="Available Cuisines",y="Cost for 2 People(approx)",palette="rocket_r",edgecolor="black")
plt.xlabel("Available Cuisines Name",fontname="Times New Roman",fontsize=10,weight="bold")
plt.ylabel("Cost for 2 People(approx)",fontname="Times New Roman",fontsize=10,weight="bold")
plt.xticks(weight="bold",rotation=85)
plt.yticks(fontsize=10,weight="bold")
for i in cu.containers:
    cu.bar_label(i)
plt.show()
```

**South Indian,North Indian,Chinese** are the only cuisines which falls under Rs 40 for two people
